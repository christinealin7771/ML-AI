```python
'''
Step 3: Acoustic Feature Extraction 
&
Step 4: Feature Post-processing
'''

def feature_extractor(audio_file):
    signal, sample_rate = librosa.load(audio_file)

    # audio feature extraction: loudness
    df_loudness = pd.DataFrame()
    S, phase = librosa.magphase(librosa.stft(signal))
    rms = librosa.feature.rms(S=S)
    # print(rms[0])
    df_loudness['Loudness'] = rms[0]
    df_loudness['Loudness'] = np.mean(df_loudness['Loudness'])
    # print(df_loudness['Loudness'])

    # audio feature extraction: mel-frequency cepstral coefficients
    df_mfccs = pd.DataFrame()
    mfccs = librosa.feature.mfcc(y=signal, sr=sample_rate, n_mfcc=12)
    for n_mfcc in range(len(mfccs)):
        df_mfccs['MFCC_%d'%(n_mfcc+1)] = mfccs[n_mfcc]
        df_mfccs['MFCC_%d'%(n_mfcc+1)] = np.mean(df_mfccs['MFCC_%d'%(n_mfcc+1)])
    # mfccs_scaled_features = np.mean(mfccs_features.T,axis=0)
    # print(df_mfccs)

    # audio feature extraction: zero crossing rate
    df_zero_crossing_rate = pd.DataFrame()
    zcr = librosa.feature.zero_crossing_rate(y=signal)
    df_zero_crossing_rate['ZCR'] = zcr[0]
    df_zero_crossing_rate['ZCR'] = np.mean(df_zero_crossing_rate['ZCR'])
    # print(df_zero_crossing_rate['ZCR'])

    # audio feature extraction: chroma
    df_chroma = pd.DataFrame()
    chromagram = librosa.feature.chroma_stft(y=signal, sr=sample_rate)
    for n_chroma in range(len(chromagram)):
        df_chroma['Chroma_%d'%(n_chroma+1)] = chromagram[n_chroma]
        df_chroma['Chroma_%d'%(n_chroma+1)] = np.mean(df_chroma['Chroma_%d'%(n_chroma+1)])
    # print(df_chroma.head(5))


    # audio feature extraction: mel spectrogram
    df_mel_spectrogram = pd.DataFrame()
    mel_spectrogram = librosa.feature.melspectrogram(y=signal, sr=sample_rate, n_mels=12)
    for n_mel in range(len(mel_spectrogram)):
        df_mel_spectrogram['Mel_Spectrogram_%d'%(n_mel+1)] = mel_spectrogram[n_mel]
        df_mel_spectrogram['Mel_Spectrogram_%d'%(n_mel+1)] = np.mean(df_mel_spectrogram['Mel_Spectrogram_%d'%(n_mel+1)])
    # print(df_mel_spectrogram.head(5))

    # combine all features
    feature_matrix = pd.concat([df_loudness, df_mfccs, df_zero_crossing_rate, df_chroma, df_mel_spectrogram], axis=1)
    # print(feature_matrix)
    # feature_matrix['mean'] = feature_matrix.mean(axis=1)
    # print(feature_matrix)

    return pd.DataFrame(feature_matrix.loc[[0]])

# Extract all the Audios
x_train_angry_df = pd.DataFrame()
for row in x_angry_train:
    x_train_angry_df = x_train_angry_df.append(feature_extractor(row), ignore_index=True)
# print(x_train_angry_df)

x_test_angry_df = pd.DataFrame()
for row in x_angry_test:
    x_test_angry_df = x_test_angry_df.append(feature_extractor(row), ignore_index=True)
# print(x_test_angry_df)

# fear_df = pd.DataFrame()
# for index, row in fear_audio_df.iterrows():
#     fear_df = fear_df.append(feature_extractor(row["audio"]), ignore_index=True)
# fear_df["emotion"] = "fear"

x_train_fear_df = pd.DataFrame()
for row in x_fear_train:
    x_train_fear_df = x_train_fear_df.append(feature_extractor(row), ignore_index=True)
# print(x_train_fear_df)

x_test_fear_df = pd.DataFrame()
for row in x_fear_test:
    x_test_fear_df = x_test_fear_df.append(feature_extractor(row), ignore_index=True)
# print(x_test_fear_df)

# happy_df = pd.DataFrame()
# for index, row in happy_audio_df.iterrows():
#     happy_df = happy_df.append(feature_extractor(row["audio"]), ignore_index=True)
# happy_df["emotion"] = "happy"

x_train_happy_df = pd.DataFrame()
for row in x_happy_train:
    x_train_happy_df = x_train_happy_df.append(feature_extractor(row), ignore_index=True)
# print(x_train_happy_df)

x_test_happy_df = pd.DataFrame()
for row in x_happy_test:
    x_test_happy_df = x_test_happy_df.append(feature_extractor(row), ignore_index=True)
# print(x_test_happy_df)

# sad_df = pd.DataFrame()
# for index, row in sad_audio_df.iterrows():
#     sad_df = sad_df.append(feature_extractor(row["audio"]), ignore_index=True)
# sad_df["emotion"] = "sad"

x_train_sad_df = pd.DataFrame()
for row in x_sad_train:
    x_train_sad_df = x_train_sad_df.append(feature_extractor(row), ignore_index=True)
# print(x_train_sad_df)

x_test_sad_df = pd.DataFrame()
for row in x_sad_test:
    x_test_sad_df = x_test_sad_df.append(feature_extractor(row), ignore_index=True)
# print(x_test_sad_df)
```

```python
'''
Step 5: Build your audio emotion recognition model
'''
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrix, roc_curve, auc, classification_report

lc = LogisticRegression()
svc = SVC(probability=True)
nbc = GaussianNB()
rfc = RandomForestClassifier()

x_train = pd.concat([x_train_angry_df, x_train_fear_df, x_train_happy_df, x_train_sad_df])
y_train = pd.concat([y_angry_train, y_fear_train, y_happy_train, y_sad_train])
print(x_train)
print(y_train)

x_test = pd.concat([x_test_angry_df, x_test_fear_df, x_test_happy_df, x_test_sad_df])
y_test = pd.concat([y_angry_test, y_fear_test, y_happy_test, y_sad_test])
print(x_test)
print(y_test)

# lc.fit(x_train, y_train)
svc.fit(x_train, y_train)
nbc.fit(x_train, y_train)
rfc.fit(x_train, y_train)
```

```python
# y_lc_predicted = lc.predict(x_test)
# print(classification_report(y_test, y_lc_predicted))

y_svc_predicted = svc.predict(x_test)
print(classification_report(y_test, y_svc_predicted))

y_nbc_predicted = nbc.predict(x_test)
print(classification_report(y_test, y_nbc_predicted))

y_rfc_predicted = rfc.predict(x_test)
print(classification_report(y_test, y_rfc_predicted))
```
